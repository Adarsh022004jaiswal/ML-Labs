```python
# Step 1: Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

# Import Pandas library
import pandas as pd

# Load the Excel dataset correctly
data = pd.read_excel(r'D:\ML_dataset3.xlsx')  # Use 'pd.read_excel' for Excel files

# Explore the data
print(data.head())
print(data.info())

# Step 3: Explore the Dataset
print(data.head())
print(data.info())

# Step 4: Preprocess the Data
# Encode categorical columns
label_encoders = {}
for column in ['Country', 'Disease Name', 'Disease Category', 'Age Group', 'Gender', 'Treatment Type', 'Availability of Vaccines/Treatment']:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le

# Normalize numerical columns
scaler = StandardScaler()
numerical_columns = ['Prevalence Rate (%)', 'Incidence Rate (%)', 'Mortality Rate (%)',
                     'Population Affected', 'Healthcare Access (%)', 'Doctors per 1000',
                     'Hospital Beds per 1000', 'Average Treatment Cost (USD)',
                     'Recovery Rate (%)', 'DALYs', 'Improvement in 5 Years (%)',
                     'Per Capita Income (USD)', 'Education Index', 'Urbanization Rate (%)']
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])

# Step 5: Define Features and Labels
X = data.drop(columns=['Disease Name'])  # Features
y = data['Disease Name']  # Target Label

# Step 6: Split the Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 7: Train the KNN Model
knn = KNeighborsClassifier(n_neighbors=5)  # You can adjust n_neighbors
knn.fit(X_train, y_train)

# Step 8: Evaluate the Model
y_pred = knn.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Optional: Test with a single row
sample = X_test.iloc[0]
prediction = knn.predict([sample])
print("Predicted Disease:", label_encoders['Disease Name'].inverse_transform(prediction))
```
